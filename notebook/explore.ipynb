```python
test_prep_mapping = {'none': 0, 'completed': 1}
lunch_mapping = {'standard': 0, 'free/reduced': 1}
gender_mapping = {'female': 0, 'male': 1}

# Aplicar el mapeo a la columna
df['test_preparation_course_encoded'] = df['test preparation course'].map(test_prep_mapping)
df['lunch_encoded'] = df['lunch'].map(lunch_mapping)
df['gender_encoded'] = df['gender'].map(gender_mapping)
```

```python
# Definir el orden de las categorías de menor a mayor nivel educativo
education_order = [
    "some high school",
    "high school",
    "some college",
    "associate's degree",
    "bachelor's degree",
    "master's degree"
]

# Crear un diccionario de mapeo
education_mapping = {category: i for i, category in enumerate(education_order)}

# Aplicar el mapeo a la columna
df['parental level of education_encoded_manual'] = df['parental level of education'].map(education_mapping)

print("Primeras filas con la codificación manual:")
print(df[['parental level of education', 'parental level of education_encoded_manual']].head())

print("\nVerificación de la codificación manual (value_counts con la codificación):")
print(df['parental level of education_encoded_manual'].value_counts().sort_index())
```

```python
df = df.drop(columns=['parental level of education']) 
df = df.drop(columns=['test preparation course'])
df = df.drop(columns=['lunch'])
df = df.drop(columns=['gender'])
```

```python
# Aplica One-Hot Encoding a 'race/ethnicity'
# usando drop_first=True para evitar la trampa de las dummies.

# Aquí incluyo solo las que deberían ser One-Hot Encoded, asumiendo las otras ya codificadas.
nominal_cols_to_encode = ['race/ethnicity']


df = pd.get_dummies(df, columns=nominal_cols_to_encode, drop_first=True)

# Verifica las nuevas columnas
print("Columnas después de One-Hot Encoding:")
print(df.columns)

print("\nPrimeras 5 filas con las variables One-Hot Encoded (ejemplo para race/ethnicity):")
print(df[['race/ethnicity_group B', 'race/ethnicity_group C',
                    'race/ethnicity_group D', 'race/ethnicity_group E']].head()) # Muestra algunas columnas dummy
```

```python
# Crear una matriz de correlación
score_corr = df[['math score', 'reading score', 'writing score']].corr()
print("\nMatriz de Correlación de Puntuaciones:")
print(score_corr)

# Visualizar la matriz de correlación con un heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(score_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor de Correlación de Puntuaciones')
plt.show()
```
